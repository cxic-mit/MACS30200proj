### 2. Load the MNIST dataset 

```python
from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# Preparing the image data
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255

# Preparing the labels
from keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

### 3. Implement a series of neural network models

### i. Initial test 

```python
from keras import models
from keras import layers

network_ini = models.Sequential()
network_ini.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network_ini.add(layers.Dense(512, activation='relu'))
network_ini.add(layers.Dense(512, activation='relu'))
network_ini.add(layers.Dense(512, activation='relu'))
network_ini.add(layers.Dense(10, activation='softmax'))

network_ini.compile(optimizer='rmsprop', 
                    loss='categorical_crossentropy', 
                    metrics=['accuracy'])

result_ini = network_ini.fit(train_images, train_labels,
                             validation_split = 1/6,
                             epochs = 200, batch_size = 512)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

val_acc = result_ini.history['val_acc']
val_loss = result_ini.history['val_loss']
plt.plot(val_acc)
plt.plot(val_loss)
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.legend(['validate_accuracy', 'validate_loss'])
plt.title('The validation set accuracy and loss over epochs')
plt.show()
```

#### Comment: 
The model's validation accuracy quickly increased to about 98% after 12 epochs, and then it keeps stable between 97% to 98.5% with small fluctuations in the following epochs. The model's validation loss firstly decreased during the first 5 epochs and then increased from the Epoch 6. Therefore, the model's performance degrades based on the validation set from the Epoch 6.

### ii. Implement dropout 

```python
network_dropout = models.Sequential()
network_dropout.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network_dropout.add(layers.Dropout(0.5))
network_dropout.add(layers.Dense(512, activation='relu'))
network_dropout.add(layers.Dropout(0.5))
network_dropout.add(layers.Dense(512, activation='relu'))
network_dropout.add(layers.Dropout(0.5))
network_dropout.add(layers.Dense(512, activation='relu'))
network_dropout.add(layers.Dropout(0.5))
network_dropout.add(layers.Dense(10, activation='softmax'))

network_dropout.compile(optimizer='rmsprop', 
                        loss='categorical_crossentropy', 
                        metrics=['accuracy'])

result_dropout = network_dropout.fit(train_images, train_labels,
                                     validation_split = 1/6,
                                     epochs = 200, batch_size = 512)
```

```python
val_acc = result_ini.history['val_acc']
val_loss = result_ini.history['val_loss']
val_acc_dropout = result_dropout.history['val_acc']
val_loss_dropout = result_dropout.history['val_loss']
plt.plot(val_acc)
plt.plot(val_loss)
plt.plot(val_acc_dropout)
plt.plot(val_loss_dropout)
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.legend(['validate_accuracy_inital', 'validate_loss_initial', 
            'validate_accuracy_dropout', 'validate_loss_dropout'])
plt.title('Compare the validation set accuracy and loss of the initial model and the dropout model over epochs')
plt.show()
```

#### Comment:
Compared to the inital model which has a rapid increase of validation loss from the Epoch 5, the dropout model has gradual increase of validation loss along the 200 epochs. The dropout model has lower validation loss than the initial model along almost all the epochs. In terms of validation accuracy, the dropout model has a little bit better accuracy than the inital model. Therefore, the dropout model may have improved performance because the intial model may be affected by overfitting.

### iii. Weight regularization 

```python
from keras import regularizers
```

```python
# Reestimate the initial model with L1 weight regularization on each layer (except the final layer) 
# with a 0.001 penalty for each weight coefficient

network_l1 = models.Sequential()
network_l1.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,), 
                            kernel_regularizer=regularizers.l1(0.001)))
network_l1.add(layers.Dense(512, activation='relu', 
                            kernel_regularizer=regularizers.l1(0.001)))
network_l1.add(layers.Dense(512, activation='relu',
                            kernel_regularizer=regularizers.l1(0.001)))
network_l1.add(layers.Dense(512, activation='relu', 
                            kernel_regularizer=regularizers.l1(0.001)))
network_l1.add(layers.Dense(10, activation='softmax'))

network_l1.compile(optimizer='rmsprop', 
                   loss='categorical_crossentropy', 
                   metrics=['accuracy'])

result_l1 = network_l1.fit(train_images, train_labels,
                           validation_split = 1/6, 
                           epochs=200, batch_size=512)
```

```python
# Reestimate the initial model with L2 weight regularization on each layer (except the final layer) 
# with a 0.001 penalty for each weight coefficient

network_l2 = models.Sequential()
network_l2.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,), 
                            kernel_regularizer=regularizers.l2(0.001)))
network_l2.add(layers.Dense(512, activation='relu', 
                            kernel_regularizer=regularizers.l2(0.001)))
network_l2.add(layers.Dense(512, activation='relu', 
                            kernel_regularizer=regularizers.l2(0.001)))
network_l2.add(layers.Dense(512, activation='relu', 
                            kernel_regularizer=regularizers.l2(0.001)))
network_l2.add(layers.Dense(10, activation='softmax'))

network_l2.compile(optimizer='rmsprop', 
                loss='categorical_crossentropy', 
                metrics=['accuracy'])

result_l2 = network_l2.fit(train_images, train_labels,
                           validation_split = 1/6,
                           epochs = 200, batch_size = 512)
```

```python
val_loss_ini = result_ini.history['val_loss']
val_loss_dropout = result_dropout.history['val_loss']
val_loss_l1 = result_l1.history['val_loss']
val_loss_l2 = result_l2.history['val_loss']
plt.plot(val_loss_ini)
plt.plot(val_loss_dropout)
plt.plot(val_loss_l1)
plt.plot(val_loss_l2)
plt.xlabel('Epochs')
plt.ylabel('Validation Loss')
plt.legend(['val_loss_initial', 'val_loss_dropout', 'val_loss_l1', 'val_loss_l2'])
plt.title('The validation loss for the initial vs. dropout vs. L1 vs.L2 model over epochs')
plt.show()
```

#### Comment:
Comparing all four models, the dropout model seems to have the lowest validation loss, so it may be the best model.
